```python
import numpy as np
import matplotlib.pyplot as plt

fx = lambda x: x**3+4*x**2+x
gx = lambda x: -(x**2) - 5*x
st = lambda x: -3*x

x = np.linspace(-10, 10, 100)
y1 = fx(x)
y2 = gx(x)
y3 = st(x)

plt.plot(x, y1, label='f(x)', color='red')
plt.plot(x, y2, label='g(x)', color='green')
plt.plot(x, y3, label='h(x)', color='blue')

plt.grid(True)
plt.xlim(-10, 10)
plt.ylim(-6, 10)

plt.legend()
plt.show()
```
