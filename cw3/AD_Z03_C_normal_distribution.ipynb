# Wielowymiarowy rozkład normalny

```python
import matplotlib.pylab as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.mlab as mlab
from matplotlib import cm
from scipy.stats import multivariate_normal
```

# Narysujmy gęstość dwuwymiarowego rozkładu Normalnego

```python
mean = np.zeros(2)
cov = np.eye(2) * 15

X1_rv=multivariate_normal(mean, cov)

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)


Z = X1_rv.pdf(np.array([X.flatten(), Y.flatten()]).T).reshape(100,100)

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(projection="3d")
ax.plot_surface(X, Y, Z, cmap="viridis")
plt.show()
```

# Narysujmy kontury gęstości dwuwymiarowego rozkładu Normalnego

```python
plt.figure(figsize=(4,4))
plt.contourf(X, Y, Z, cmap="viridis")
plt.show()
```

## Narysujmy poziomice (contour plot) dla mieszanki gaussów

Rozpatrzmy mieszankę gaussów postaci
$$
\frac{1}{2}N(m_1,\Sigma_1)+\frac{1}{2}N(m_2,\Sigma_2).
$$
Proszę narysować poziomice dla przykładu gdy $m_1\neq m_2$, $\Sigma_1=\Sigma_2$

```python
z1 = multivariate_normal([0,0], np.eye(2)).pdf(np.array([X.flatten(), Y.flatten()]).T).reshape(100,100)
z2 = multivariate_normal([3,3], np.eye(2)).pdf(np.array([X.flatten(), Y.flatten()]).T).reshape(100,100)

z = 1/2 * z1 + 1/2 * z2

plt.figure(figsize=(7,7))
plt.contourf(X, Y, z, cmap="viridis")
plt.show()
```

# Wylosujmy próbkę z dwuwymiarowego rozkładu Normalnego

Proszę wylosować na dwa sposoby: korzystając z scipy.stats.multivariate_normal oraz numpy.random.multivariate_normal

```python
sample_scipy = multivariate_normal(np.zeros(2), np.eye(2)).rvs(1000)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6,6))
ax.scatter(sample_scipy[:, 0], sample_scipy[:, 1])
ax.set_xlim([-3,3])
ax.set_ylim([-3,3])
plt.show()
```

```python
sample_scipy = np.random.multivariate_normal(np.zeros(2), np.eye(2), size=1000)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6,6))
ax.scatter(sample_scipy[:, 0], sample_scipy[:, 1])
ax.set_xlim([-3,3])
ax.set_ylim([-3,3])
plt.show()
```
