# Ridge Regression

```python
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from sklearn import linear_model
import scipy.stats as stats
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn import model_selection
from sklearn.model_selection import GridSearchCV
```

Rozważmy następujący zbiór punktów:

```python
true_fun = lambda X: np.cos(1.5 * np.pi * X)
n_samples = 20
x = np.sort(np.random.rand(n_samples))
y = true_fun(x) + np.random.randn(n_samples) * 0.1
x = np.vstack(x)
```

# Przykład 
Proszę wykonać regresję (Ridge Regression) na powyższym zbiorze danych 

```python
clf = Ridge(alpha=1.0)
clf.fit(x, y)

x_plot = np.vstack(np.linspace(0, 1, 20))
plt.plot(x_plot, clf.predict(x_plot), color='blue', linewidth=3)
plt.plot(x, y, 'ok');
plt.show()
```

## Regresja liniowa

```python
# prepare models
models = []
predicts = []
names = []
models.append(('LR degree 2', make_pipeline(PolynomialFeatures(2), linear_model.LinearRegression())))
models.append(('LR degree 20', make_pipeline(PolynomialFeatures(20), linear_model.LinearRegression())))

x_plot = np.vstack(np.linspace(-3, 3, 1000))
for name, model in models:
    print(name)
    model.fit(x, y)
    predicts.append(model.predict(x_plot))
    names.append(name)

x_plot = np.vstack(np.linspace(-3, 3, 1000))
plt.plot(x, y, 'ok');
for i in range(len(models)):
    #print(i)
    plt.plot(x_plot, predicts[i], linewidth=3, label=names[i])
    plt.xlim((0, 1))
    plt.ylim((-2, 2))
plt.legend()
plt.show()
```

# Zadanie 
Proszę wykonać 
* regresję dla wielomianów o stopniu **20**
* regresję (Ridge Regression) dla wielomianów o stopniu **20** oraz
   * alpha = 1 
   * alpha = 10 000 
   * alpha = 0.0001


```python
# prepare models
models = []
models.append(('LR', make_pipeline(PolynomialFeatures(20), linear_model.LinearRegression())))
models.append(('RR α=1', make_pipeline(PolynomialFeatures(20), Ridge(alpha=1))))
models.append(('RR α=10 000', make_pipeline(PolynomialFeatures(20), Ridge(alpha=10_000))))
models.append(('RR α=0.0001', make_pipeline(PolynomialFeatures(20), Ridge(alpha=0.0001))))

predicts = []
names = []
x_plot = np.vstack(np.linspace(-3, 3, 1000))

for name, model in models:
    model.fit(x, y)
    predicts.append(model.predict(x_plot))
    names.append(name)

# drawing
plt.plot(x, y, 'ok');
for i in range(len(models)):
    plt.plot(x_plot, predicts[i], linewidth=1, label=names[i])
    plt.xlim((0, 1))
    plt.ylim((-2, 2))

plt.title('Comparison of 20th degree models')
plt.legend()
plt.show()
```

# Zdanie 
Dobierz optymalny stopień wielomianu oraz parametr alpha za pomocą komendy **GridSearchCV**. 

```python
def best_fit(x, y) -> tuple[Pipeline, int, float]:
    grid = GridSearchCV(
        make_pipeline(PolynomialFeatures(), Ridge()),
        param_grid={
            "polynomialfeatures__degree": range(1, 21),
            "ridge__alpha": [0.0001, 1, 10, 100, 1000]
        },
        cv=10,
        scoring="neg_mean_squared_error",
    )
    grid.fit(x,y)
    return grid.best_estimator_, grid.best_params_["polynomialfeatures__degree"], grid.best_params_["ridge__alpha"]
```

```python
model, deg, alpha = best_fit(x, y)

x_plot = np.vstack(np.linspace(-3, 3, 1000))
plt.plot(x, y, 'ok');

plt.plot(x_plot, model.predict(x_plot), linewidth=1, label=f'RR deg={deg}, α={alpha:.4}')
plt.xlim((0, 1))
plt.ylim((-2, 2))
plt.legend()

plt.show()
```
