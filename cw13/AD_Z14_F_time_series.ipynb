```python
import pandas
import matplotlib.pyplot as plt
dataframe = pandas.read_csv('Dane/monthly-car-sales-in-quebec-1960.csv', sep=";")
dataframe.columns = ['Month', 'Sales']

# Convert 'Month' to datetime if needed
dataframe['Month'] = pandas.to_datetime(dataframe['Month'], format='%Y-%m')

# Plot the data
plt.plot(dataframe['Month'], dataframe['Sales'])
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Car Sales in Quebec (1960)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
